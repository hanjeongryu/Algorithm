## 나의 풀이
[풀이 key]
- DFS 활용
- DFS 코드
  - for문으로 방향 횟수만큼 돌기  
  - 더이상 연결할 0 부분이 없을 때 1을 return


```python
import time

##################
##### 인자 받기
##################
# 행,열
n,m = map(int,input().split())
# 매트릭스
mat = []
for n_i in range(n):
  mat.append(list(map(int,input())))

start_time = time.time()

##################
##### 움직일 경로 설정
##################
# 상우하좌
dx = [1,0,-1,0]
dy = [0,1,0,-1]

##################
##### dfs 함수 만들기
##################
def dfs(x,y):
  mat[x][y] = 1
  # 상우하좌로 연결된 노드 탐색 시작
  for i in range(4):
    nx = x + dx[i]
    ny = y + dy[i]
    # 주어진 범위 안에 있는 경우만 탐색 계속
    if nx >= 0 and nx < n and ny >= 0 and ny < m:
      # 새롭게 탐색하는 음료수(=미방문) 부분이 있을 때 방문(=얼음=칸막이) 처리
      if mat[nx][ny] == 0:
        mat[nx][ny] = 1
        dfs(nx,ny)
  return 1

##################
##### 매트릭스 탐색
##################
result = 0
for n_i in range(n):
  for m_i in range(m):
    if mat[n_i][m_i] == 0:
      # 현재 위치에서 dfs를 수행(한 덩어리의 음료수를 모두 방문 처리한 후 result에 +1)
      result += dfs(n_i,m_i)

print(result)
end_time = time.time()
print("소요시간: ", end_time - start_time)
```

## 정답지
[풀이 key]
- DFS 활용
- DFS 코드
  - for문 없이 방향 횟수만큼 dfs 반복 사용
  - 조건에 해당하지 않을 시 False, 해당할 시 True 리턴


```python
import time

##################
##### 인자 받기
##################
# 행,열
n,m = map(int,input().split())
# 매트릭스
graph = []
for n_i in range(n):
  graph.append(list(map(int,input())))

start_time = time.time()

##################
##### dfs 만들기
##################
def dfs(x,y):
  if x <= -1 or x >= n or y <= -1 or y >= m:
    return False
  if graph[x][y] == 0:
    # 해당 노드 방문 처리
    graph[x][y] = 1
    # 상하좌우의 위치들도 모두 재귀적으로 호출
    print("########## ({},{}) 탐색 시작 ##########".format(x,y))
    print("({},{})의 상".format(x,y))
    dfs(x-1,y)
    print("({},{})의 하".format(x,y))
    dfs(x+1,y)
    print("({},{})의 좌".format(x,y))
    dfs(x,y-1)
    print("({},{})의 우".format(x,y))
    print("========== ({},{}) 탐색 종료 ==========".format(x,y))
    dfs(x,y+1)
    return True
  return False

##################
##### 매트릭스 탐색
##################
result = 0
for i in range(n):
  for j in range(m):
    if dfs(i,j) == True:
      result += 1

print(result)
end_time = time.time()
print("소요시간: ", end_time - start_time)
```

## 정답지
[풀이 key]
- DFS 활용
- DFS 코드
  - for문으로 방향 횟수만큼 돌기  
  - 조건에 해당하지 않을 시 False, 해당할 시 True 리턴


```python
import time

##################
##### 인자 받기
##################
# 행,열
n,m = map(int,input().split())
# 매트릭스
graph = []
for n_i in range(n):
  graph.append(list(map(int,input())))

start_time = time.time()

##################
##### dfs 만들기
##################
def dfs(x,y):
  if x <= -1 or x >= n or y <= -1 or y >= m:
    return False
  if graph[x][y] == 0:
    # 해당 노드 방문 처리
    graph[x][y] = 1
    # 상하좌우의 위치들도 모두 재귀적으로 호출
    print("########## ({},{}) 탐색 시작 ##########".format(x,y))
    print("({},{})의 상".format(x,y))
    dfs(x-1,y)
    print("({},{})의 하".format(x,y))
    dfs(x+1,y)
    print("({},{})의 좌".format(x,y))
    dfs(x,y-1)
    print("({},{})의 우".format(x,y))
    print("========== ({},{}) 탐색 종료 ==========".format(x,y))
    dfs(x,y+1)
    return True
  return False

##################
##### 매트릭스 탐색
##################
result = 0
for i in range(n):
  for j in range(m):
    if dfs(i,j) == True:
      result += 1

print(result)
end_time = time.time()
print("소요시간: ", end_time - start_time)
```

## 정답지와 나의 풀이 비교
- 정답지 풀이와 비교했을 때 dfs 함수 내 for문 사용 여부와 return 사용 방식에 차이가 있음  
  - for문 사용 여부에 대한 차이:
내 풀이는 for문을 사용하는 반면 정답지는 for문 대신 dfs문을 반복 서술하는 방식
  - return 사용 방식에 대한 차이:  
내 풀이는 더이상 연결된 1 부분이 없을 때만 return을 하는 반면 정답지는 조건을 벗어날 때는 False, 조건에 해당할 때는 True를 return하는 식
- 속도
  - 내 풀이 방식이 약 2배 빠름. 리턴 횟수에 따른 차이로 추정됨
